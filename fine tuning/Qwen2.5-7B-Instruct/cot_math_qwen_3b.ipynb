```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 8192
dtype = None
load_in_4bit = True

fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit",
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",
]

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen2.5-3B",
    max_seq_length = 8192,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 32,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
```

```python
# Old:
# from datasets import load_dataset
# dataset = load_dataset("Ashkchamp/Openthoughts_math_filtered_30K", split="train")

import kagglehub
from kagglehub import KaggleDatasetAdapter
from datasets import Dataset      # HF datasets
# optional: import pandas as pd   # not needed if kagglehub already returns DataFrame

# Name of the Excel file in the Kaggle dataset
train_file = "gsm8k_train_translated_all.xlsx"

# Load as pandas DataFrame
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "arefehrajabian/persian-gsm",
    train_file,
)

# (Optional) keep / rename the columns to match what your training code expects
# For example, if your old dataset had "question" and "answer" fields:
df = df[["question", "answer", "steps", "translated_question", "translated_steps"]]

# Convert pandas DataFrame -> HuggingFace Dataset
dataset = Dataset.from_pandas(df, preserve_index=False)

# Now `dataset` behaves like the old HF dataset object
print(dataset)
print(dataset[0])
```

```python
EOS_TOKEN = tokenizer.eos_token

def create_prompt(example, train=False):
    """
    Adapted for the Persian GSM8K dataset.
    Uses:
      - translated_question (fallback to question)
      - translated_steps (fallback to steps)
      - answer
    """

    # You can hardcode a system message or leave it empty
    system = "You are a helpful math tutor."

    # Prefer Persian translation; fall back to English if missing
    question = str(
        example.get("translated_question")
        or example.get("question", "")
    )

    # Chain-of-thought: again, prefer translated_steps
    cot = str(
        example.get("translated_steps")
        or example.get("steps", "")
    )

    solution = str(example.get("answer", ""))

    if train:
        # For training you usually want the full target (CoT + final answer)
        prompt = (
            f"System: {system}\n"
            f"Question: {question}\n"
            "Let's reason step by step:\n"
            f"{cot}\n"
            "Therefore, the final answer is:\n"
            f"{solution}<|final|>{EOS_TOKEN}"
        )
    else:
        # For inference you usually only give the question + instruction
        prompt = (
            f"System: {system}\n"
            f"Question: {question}\n"
            "Let's reason step by step:\n"
        )

    return prompt
```

```python
def tokenize_function(batch, train=True):
    # Build prompts for each example in the batch
    prompts = []

    # Use "answer" to determine batch size (present in every row)
    batch_size = len(batch["answer"])

    for i in range(batch_size):
        ex = {
            # Prefer translated fields if present, otherwise fall back
            "translated_question": batch["translated_question"][i] if "translated_question" in batch else "",
            "question": batch["question"][i] if "question" in batch else "",
            "translated_steps": batch["translated_steps"][i] if "translated_steps" in batch else "",
            "steps": batch["steps"][i] if "steps" in batch else "",
            "answer": batch["answer"][i],
        }

        prompts.append(create_prompt(ex, train=train))

    tokenized = tokenizer(
        prompts,
        truncation=True,
        max_length=8192,
        padding="max_length",
    )

    # Required by SFTTrainer (or similar): keep raw text
    tokenized["text"] = prompts
    return tokenized
```

```python
tokenized_train = dataset.map(tokenize_function, batched=True, remove_columns=dataset.column_names)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments, DataCollatorForLanguageModeling
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=tokenized_train,
    dataset_text_field="text",
    max_seq_length=8192,
    dataset_num_proc=2,
    packing=False,
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False),
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        warmup_steps=5,
        # num_train_epochs = 1,
        max_steps=120,
        learning_rate=2e-5,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
        report_to="none",
    ),
)
```

```python
import gc
gc.collect()
torch.cuda.empty_cache()
```

```python
trainer.train()
```

```python
from IPython.display import Markdown, display

# Enable native 2x faster inference
FastLanguageModel.for_inference(model)

# Prepare an example for inference.
# For inference, leave 'COT_Reason' and 'solution' empty so the model generates them.
example = {
    "system": (
        "Your role as an assistant involves thoroughly exploring questions through a systematic long thinking process before providing the final precise and accurate solutions. "
        "This requires engaging in a comprehensive cycle of analysis, summarizing, exploration, reassessment, reflection, backtracing, and iteration to develop a well-considered thinking process. "
        "Please structure your response into two main sections: Thought and Solution. In the Thought section, detail your reasoning process using the specified format: "
        "<|begin_of_thought|> {thought with steps separated with '\\n\\n'} <|end_of_thought|> "
        "Each step should include detailed considerations such as analyzing questions, summarizing relevant findings, brainstorming new ideas, verifying the accuracy of the current steps, refining any errors, and revisiting previous steps. "
        "In the Solution section, based on various attempts, explorations, and reflections from the Thought section, systematically present the final solution that you deem correct. "
        "The solution should remain a logical, accurate, concise expression style and detail the necessary steps needed to reach the conclusion, formatted as follows: "
        "<|begin_of_solution|> {final formatted, precise, and clear solution} <|end_of_solution|> "
        "Now, try to solve the following question through the above guidelines."
    ),
    "problem": (
        "Emily buys the following three items from an online store:\n\n"
        "A shirt for $12.45\n"
        "A pair of socks for $7.99\n"
        "A sweater for $25.10\n\n"
        "She applies a coupon that gives 10% off the total cost of these items. After the discount, an 8% sales tax is applied to the discounted total. "
        "Finally, there is a $3.00 shipping fee added to the purchase.\n\n"
        "Question: What is Emilyâ€™s final total amount (to the nearest cent) after the discount, sales tax, and shipping fee?"
    ),
    "COT_Reason": "",
    "solution": ""
}

# Create the prompt using the new template.
# Use train=True so that the prompt ends with "Let's reason step by step:".
prompt = create_prompt(example, train=True)

# Tokenize the prompt.
inputs = tokenizer([prompt], return_tensors="pt").to("cuda")

# Generate output from the model.
outputs = model.generate(**inputs, max_new_tokens=1024, use_cache=True)

# Decode the generated tokens.
result = tokenizer.batch_decode(outputs)

# Join the output with newlines for better Markdown formatting.
markdown_output = "\n\n".join(result)

# Display the result as Markdown.
display(Markdown(markdown_output))
```
